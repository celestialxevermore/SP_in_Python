2. 가설 수립

선형 회귀의 가설 : 1차원 방정식 + y절편 

H(x) = Wx + b



파이토치로 선형 회귀 구현하기

1. 기본 세팅

```python
import torch 
import torch.nn as nn 
import torch.nn.functional as F 
import torch.optim as optim 
```

```python
torch.manual_seed(1)
```

2. 변수 선언

```python
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[2],[4],[6]])
```

```python
print(x_train.shape)
print(y_train.shape)
```

3. 가중치와 편향의 초기화 

선형 회귀란, 학습 데이터와 가장 잘 맞는 하나의 직선을 긋는 일.

그 직선을 정의 하는 W,b를 구하기.

```python
Weight = torch.zeros(1,requires_grad=True)
print(Weight)
```

가중치 Weight는 0으로 초기화된, 크기 1짜리 1차원 벡터입니다. 

requires_grad = True : 변수가 학습을 통해 계속 변화

```python
bias  = torch.zeros(1,requires_grad=True)
print(bias)
```

4. 가설 세우기

- 파이토치 코드 상으로 직선의 방정식에 해당되는 가설을 선언합니다.

```python
hypothesis = x_train * Weight + bias 
print(hypothesis)
print(hypothesis.shape)
```

5. 비용 함수 선언하기 

```python
cost = torch.mean((hypothesis-y_train)**2)
print(cost)
```

6. 경사 하강법 구현

SGD : Stochastic Gradient Descent 

```python
optimizer = optim.SGD([Weight,bias],lr=0.01)
```

```python
optimizer.zero_grad()

cost.backward()

optimizer.step()
```

7. 전체 코드 

```python
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[2],[4],[6]])


# 모델 초기화
W = torch.zeros(1,requires_grad=True)
b = torch.zeros(1,requires_grad=True)

#optimizer 설정 
optimizer = optim.SGD([W,b],lr=0.01)

nb_epochs = 1999


for epoch in range(nb_epochs+1):

    #H(x)
    hypothesis = x_train * W + b

    #cost 계산 
    cost = torch.mean((hypothesis-y_train)**2) 

    #cost로 H(x)개선 
    optimizer.zero_grad()
    #역전파
    cost.backward()
    #진행
    optimizer.step()

    #100번마다 로그 출력
    if epoch%100 ==0:
        print('Epoch {:4d}/{} W: {:.3f}, b : {:.3f} Cost : {:.6f}'.format(epoch,nb_epochs,W.item(),b.item(),cost.item()))
```
